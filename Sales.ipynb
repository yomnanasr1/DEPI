# The total revenue generated from all orders

```python
df['Revenue'] = df['Quantity'] * df['Price']
total_revenue = df['Revenue'].sum()
print(f"Total Revenue: {total_revenue}")
```

# The average amount of money spent per order

```python
total_orders = df['Order ID'].nunique()
average_per_order = round(total_revenue / total_orders)
print(f"Average Amount Spent Per Order: {average_per_order}")
```

# The total revenue by category

```python
total_revenue_per_category = df.groupby('Category')['Revenue'].sum()

plt.figure(figsize=(8, 6))
total_revenue_per_category.sort_values(ascending=False).plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Total Revenue per Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)

# Show the chart
plt.show()
```

# The total revenue by Product

```python
plt.figure(figsize=(25, 8))
sns.heatmap(df[['Product', 'Revenue']].set_index('Product').T, cmap='Blues', annot=False, cbar=True)

plt.title('Revenue for All Products', fontsize=14)

plt.show()
```

#  Top 10 Cities by Revenue

```python
sorted_revenue = df.groupby('City')['Revenue'].sum().sort_values(ascending=False)

# Select top 10 cities by revenue
top_cities = sorted_revenue.head(10)

# Plot the top 10 cities by revenue
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Top 10 Cities by Revenue', fontsize=14)
plt.xlabel('City', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

# The total revenue by Country

```python
total_revenue_per_country = df.groupby('Country')['Revenue'].sum()

# Sort by revenue in descending order
total_revenue_per_country = total_revenue_per_country.sort_values(ascending=False)
# Plot total revenue per country as a horizontal bar chart

plt.figure(figsize=(12, 10))
total_revenue_per_country.plot(kind='barh', color='lightcoral')

# Add labels and title
plt.title('Total Revenue per Country', fontsize=14)
plt.xlabel('Total Revenue', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Show the chart
plt.tight_layout()
plt.show()
```

# The top 5 cities with the highest sales revenue In USA

```python
usa = df[df['Country'] == 'USA']
# Group by city and sum the revenue
city_revenue = usa.groupby('City')['Revenue'].sum().reset_index()
# Sort by revenue in descending order and get the top 5 cities
top_5_cities = city_revenue.sort_values(by='Revenue', ascending=False).head(5)
# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_cities, x='City', y='Revenue')
# Add titles and labels
plt.title('Top 5 Cities with Highest Sales Revenue in the USA', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

# Top 10  Total Revenue by Customer

```python
total_revenue_per_customer = df.groupby('Customer')['Revenue'].sum()

# Sort by revenue in descending order (optional)
total_revenue_per_customer = total_revenue_per_customer.sort_values(ascending=False)

top_10_customers = total_revenue_per_customer.head(10)

plt.figure(figsize=(10, 6))
top_10_customers.plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Total Revenue by Customer', fontsize=14)
plt.xlabel('Customer Name', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)

plt.show()
```

# Total Revenue by Shipping Company

```python
# Calculate total revenue for each shipping company
total_revenue_per_shipper = df.groupby('Shipper')['Revenue'].sum()

# Sort by revenue in descending order
total_revenue_per_shipper = total_revenue_per_shipper.sort_values(ascending=False)
custom_colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']
plt.figure(figsize=(8, 8))
plt.pie(total_revenue_per_shipper, labels=total_revenue_per_shipper.index, autopct='%1.1f%%', startangle=90,colors=custom_colors)

# Add a title
plt.title('Total Revenue by Shipping Company', fontsize=14)

plt.show()
```

# Top 10 Total Revenue by Supplier

```python
total_revenue_per_supplier = df.groupby('Supplier')['Revenue'].sum()

# Sort by revenue in descending order (optional)
total_revenue_per_supplier = total_revenue_per_supplier.sort_values(ascending=False)

top_10_supplier = total_revenue_per_supplier.head(10)

plt.figure(figsize=(10, 6))
top_10_supplier.plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Total Revenue by Supplier', fontsize=14)
plt.xlabel('Supplier', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)

plt.show()
```

# Top 20 Best-Selling Products

```python
total_units_sold_per_product = df.groupby('Product')['Quantity'].sum()
total_units_sold_per_product = total_units_sold_per_product.sort_values(ascending=False).head(20)

plt.figure(figsize=(12, 8))
total_units_sold_per_product.plot(kind='bar', color='lightcoral')

# Add labels and title
plt.title('Top 20 Best-Selling Products', fontsize=16)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Total Units Sold', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the chart
plt.tight_layout()
plt.show()
```

# Top 5 Customers by Number of Orders

```python
orders_per_customer = df.groupby('Customer')['Order ID'].count()
orders_per_customer = orders_per_customer.reset_index(name='NumberOfOrders')
top_5_customers = orders_per_customer.sort_values(by='NumberOfOrders', ascending=False).head(5)
custom_colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

plt.figure(figsize=(8, 8))
plt.pie(top_5_customers['NumberOfOrders'], labels=top_5_customers['Customer'], autopct='%1.1f%%', startangle=140,colors=custom_colors)


# Add title
plt.title('Top 5 Customers by Number of Orders', fontsize=16)

# Show the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.show()
```
